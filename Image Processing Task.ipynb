```python

from PIL import Image
 
import numpy as np
import matplotlib.pyplot as plt

img1 = plt.imread("C:/Users/Aashu/Downloads/input_img.jpeg")
img1=Image.fromarray(np.fliplr(img1))
img1=img1.crop((830,283,1900,1200))
img2= plt.imread("C:/Users/Aashu/Downloads/input_img2.jpeg")
img2=Image.fromarray(img2)

image1 = img1.resize((500, 500))
image2 = img2.resize((500, 500))


image3 = image1.convert("RGBA")
image4 = image2.convert("RGBA")




alphaBlended1 = Image.blend(image3, image4, alpha=.2)
alphaBlended2 = Image.blend(image3, image4, alpha=.4)

# Display the alpha-blended images
alphaBlended1.show()
alphaBlended2.show()
```
